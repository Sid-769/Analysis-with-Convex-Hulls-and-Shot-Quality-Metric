```python
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.patches import Arc


def createPitch():
    fig, ax = plt.subplots(figsize=(8,5))

    # Pitch Outline & Centre Line
    ax.plot([0,0],[0,90], color="black")
    ax.plot([0,130],[90,90], color="black")
    ax.plot([130,130],[90,0], color="black")
    ax.plot([130,0],[0,0], color="black")
    ax.plot([65,65],[0,90], color="black")

    # Penalty areas and goals
    ax.plot([17,17],[75,15],color="black")
    ax.plot([0,17],[75,75],color="black")
    ax.plot([17,0],[15,15],color="black")
    ax.plot([130,113],[75,75],color="black")
    ax.plot([113,113],[75,15],color="black")
    ax.plot([113,130],[15,15],color="black")

    # 6-yard Boxes
    ax.plot([0,6],[60,60],color="black")
    ax.plot([6,6],[60,30],color="black")
    ax.plot([6,0.5],[30,30],color="black")
    ax.plot([130,124],[60,60],color="black")
    ax.plot([124,124],[60,30],color="black")
    ax.plot([124,130],[30,30],color="black")

    # Circles and Spots
    centreCircle = plt.Circle((65,45), 9.15, color="black", fill=False)
    centreSpot = plt.Circle((65,45), 0.37, color="black")
    leftPenSpot = plt.Circle((11,45), 0.37, color="black")
    rightPenSpot = plt.Circle((119,45), 0.37, color="black")
    ax.add_patch(centreCircle)
    ax.add_patch(centreSpot)
    ax.add_patch(leftPenSpot)
    ax.add_patch(rightPenSpot)

    # Arcs
    leftArc = Arc((11,45),height=18.3,width=18.3,angle=0,theta1=308,theta2=52,color="black")
    rightArc = Arc((119,45),height=18.3,width=18.3,angle=0,theta1=128,theta2=232,color="black")
    ax.add_patch(leftArc)
    ax.add_patch(rightArc)
    
    return fig, ax

fig, ax = createPitch()

plotted_points = []

def onclick(event):
    if event.xdata > 0 and event.xdata < 130 and event.ydata > 0 and event.ydata < 90:
        
        # Check for existing points near the clicked location
        to_remove = None
        for point, text in plotted_points:
            if abs(point.get_xdata()[0] - event.xdata) < 1 and abs(point.get_ydata()[0] - event.ydata) < 1:
                point.remove()
                text.remove()
                to_remove = (point, text)
                break
                
        # If we didn't find an existing point, plot a new one
        if to_remove is None:
            point, = ax.plot(event.xdata, event.ydata, 'ro')  # Plots a red point where you clicked
            text = ax.text(event.xdata + 1, event.ydata + 1, f"({event.xdata:.2f}, {event.ydata:.2f})", fontsize=8)
            plotted_points.append((point, text))
            print(f"Shot coordinates: X = {event.xdata:.2f}, Y = {event.ydata:.2f}")
        else:
            plotted_points.remove(to_remove)
        
        fig.canvas.draw()  # Update the plot

cid = fig.canvas.mpl_connect('button_press_event', onclick)
plt.show()
```
